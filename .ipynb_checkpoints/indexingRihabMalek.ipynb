# Helper functions for visualization

```python
def display_hits(res):
    print(f"Found {res['hits']['total']['value']} hits in {res['took']} ms. Showing top {len(res['hits']['hits'])}.")
    print("")
    
    for hit in res['hits']['hits']:
        
        
        s = hit['_source']
        
        print(f"ID      {s.get('id', None)}")
        print(f"Score   {hit.get('_score', None)}")
        display(Image(s.get("imgPATH"), width=128))
        print("")
```

```python
def display_hits_horizontal(res):
    print(f"Found {res['hits']['total']['value']} hits in {res['took']} ms. Showing top {len(res['hits']['hits'])}.")
    html = "<table><tr>"
    for hit in res['hits']['hits']:
        s = hit['_source']
        html += f"<td><img src=\"{s.get('imgPATH')}\" width=128 /><p>{s.get('id')}, {hit.get('_score')}</p></td>"
    html += "</tr></table>"
    display(HTML(html))
```

# Search by image similarities using approximate nearest neighbors query

```python
def search_by_image_query(image_id=None, feature_vector=None,size=10):
    if image_id==None and feature_vector==None:
        raise ValueError("Please enter an Image ID or a Feature Vector")
    
    if image_id:
        query = {
            "query": {
                "elastiknn_nearest_neighbors": {
                  "vec": {
                    "index": index,
                    "field": "featureVec",
                    "id": image_id
                  },
                  "field": "featureVec",
                  "model": "lsh",
                  "similarity": "l2",
                  "candidates": 500
                }
            }
          }
    elif feature_vector:
        query = {
            "query": {
                    "elastiknn_nearest_neighbors": {
                      "vec": {
                        "values": feature_vector
                      },
                      "field": "featureVec",
                      "model": "lsh",
                      "similarity": "l2",
                      "candidates": 150
                    }
          }
        }
    res = es.search(index=index, body=query, size=size, _source=source_no_vecs)
    display_hits(res)
```
